**Analyze the text reviews too identify the mostcommon positive and negative keywords**

```python
rating_counts = df["Rating text"].value_counts()
```

```python
#Visualize the distribution of rating text categories

plt.figure(figsize=(10,6))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.xticks(rotation=45)
plt.xlabel("Rating Text")
plt.ylabel("Count")
plt.title("Distribution of Rating Text Categories")
plt.show()
```

```python
#display the most common rating text categories

rating_counts
```

```python
from collections import Counter
```

```python
#Define positive and and negative ratings

positive_rating = ["Excellent","very Good","Good"]
negative_rating = ["Average","Poor"]
```

```python
#Count Occurrence of positive and negative words

positive_words =" ".join(df[df["Rating text"].isin(positive_rating)]["Rating text"]).lower().split()
negative_words =" ".join(df[df["Rating text"].isin(negative_rating)]["Rating text"]).lower().split()
```

```python
#Get most common words

positive_word_counts = Counter(positive_words).most_common(5)
negative_words_counts = Counter(negative_words).most_common(5)

positive_word_counts ,negative_words_counts
```

**Calculate the average length of reviews andexplore if there is a relationship between revie length and rating.**

```python
#Calculate the  length of rating text

df["Rating text length"] = df["Rating text"].apply(lambda x: len(x.split()))
```

```python
#Compute the average length of rating text

aveage_length = df["Rating text length"].mean()
```

```python
#Visualize the relationship between rating text length and aggregate rating

plt.figure(figsize=(10,6))
sns.boxplot(x="Rating text", y="Rating text length",data=df,order=rating_counts.index)
plt.xticks(rotation=45)
plt.xlabel("Rating category")
plt.ylabel("Rating text length")
plt.title("Relationship between rating text length and rating Category")
plt.show()

aveage_length
```

**Task 4 Online Delivery**

**determine the Percentage of resturants that offers Online delivery**

```python
total_resturants = len(df)
```

```python
#Determine the percentage of resturants that offer online delivery

resturants_with_online_delivery = df['Has Online delivery'].value_counts().sort_index().get('Yes')
percentage_with_online_delivery = (resturants_with_online_delivery / total_resturants) * 100
print("The Percentage of resturants that offers Online delivery",percentage_with_online_delivery)
```

**The Average Rating of resturants with and without online delivery**

```python
#Compare the average rating of resturants with and without online delivery

average_rating_with_online_delivery = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
average_rating_without_online_delivery = df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()

print("Average Rating of resturants with Online delivery is", average_rating_with_online_delivery)
print("Average Rating of resturants without Online delivery is",average_rating_without_online_delivery)
```

```python
#Create Bar chart

labels = ['With Online Delivery', 'Without Online Delivery']
average_rating = [average_rating_with_online_delivery, average_rating_without_online_delivery]

colors = ['green', 'red']

plt.bar(labels, average_rating, color=colors)
plt.title("the Comparison of Average Ratings of Resturants with and without Online Delivery")
plt.xlabel("Delivery Status")
plt.ylabel("Average Rating")
plt.show()
```
